```python
def piramide7(num, desloc=True, diamond=False):
    for c in range(1, num + 1):
        for d in range(num + desloc, 0, -1):
            if d > c:
                print(f"{'  ' * diamond}", end='')
            else:
                print(f"{'0' if c < 10 else ''}{c}", end='  ')
        print('')
```

```python
def piramide8(num, desloc=True, diamond=False):
#   for c in range(1, num + 1):
    for c in range(num, 0, -1):
        for d in range(num + desloc, 0, -1):
            if d > c:
                print(f"{'  ' * diamond}", end='')
            else:
                print(f"{'0' if c < 10 else ''}{c}", end='  ')
        print('')
```
